```python
countries = ['india', 'UK', 'egypt', 'japan', 'norway']
exports = [78, 76, 81, 84, 75]
targets = [84, 84, 88, 90, 78]
```

```python
import numpy as np
```

```python
index = np.arange(5)
w = 0.3
```

```python

plt.bar(index, exports, w)
plt.bar(index+w, targets, w)

plt.title("overall exports")
plt.xlabel("countries")
plt.ylabel("numbers")

plt.show()
```

```python

index = np.arange(5)
w = 0.3

plt.bar(index, exports, w)
plt.bar(index+w, targets, w)

plt.title("overall exports")
plt.xlabel("countries")
plt.ylabel("numbers")

plt.xticks(index, countries)

plt.show()
```

```python

index = np.arange(5)
w = 0.3

plt.bar(index, exports, w)
plt.bar(index+w, targets, w)

plt.title("overall exports")
plt.xlabel("countries")
plt.ylabel("numbers")

plt.xticks(index+0.15, countries)

plt.show()
```

```python
index = np.arange(5)
w = 0.3

plt.bar(index, exports, w, label="actual exports")
plt.bar(index+w, targets, w, label="2020 targets")

plt.title("overall exports")
plt.xlabel("countries")
plt.ylabel("numbers")

plt.ylim(0, 100)

plt.xticks(index+0.15, countries)
plt.legend()

plt.show()
```

```python
index = np.arange(5)
w = 0.3

plt.bar(index, exports, w, label="actual exports")
plt.bar(index+w, targets, w, label="2020 targets")

plt.title("overall exports")
plt.xlabel("countries")
plt.ylabel("numbers")

plt.ylim(0, 100)

plt.xticks(index+0.15, countries)
plt.legend(loc='upper left')

plt.show()
```

```python
index = np.arange(5)
w = 0.3

plt.bar(index, exports, w, label="actual exports")
plt.bar(index+w, targets, w, label="2020 targets")

plt.title("overall exports")
plt.xlabel("countries")
plt.ylabel("numbers")

plt.ylim(60, 100)

plt.xticks(index+0.15, countries)
plt.legend(loc='upper left')

plt.show()
```
